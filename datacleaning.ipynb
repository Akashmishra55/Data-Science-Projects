```python
#2nd method
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv("student_performance_prediction.csv")
data.head(10)
```

```python
from sklearn.preprocessing import FunctionTransformer
```

```python
ft=FunctionTransformer(func=lambda x:x**2)
ft.fit(data[["Attendance Rate"]])
```

```python
data["new_ft"]=ft.transform(data[["Attendance Rate"]])
data["new_ft"]
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
sns.distplot(data["Attendance Rate"])
plt.title("before")
plt.subplot(1,2,2)
sns.distplot(data["new_ft"])
plt.title("after")
plt.show()
```
