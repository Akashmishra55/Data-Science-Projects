```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dtset=pd.read_csv("Student_Performance.csv")
dtset.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
from sklearn.model_selection import train_test_split
```

```python
plt.figure(figsize=(10,5))
sns.heatmap(data=dtset.corr(),annot=True)
plt.show()           
```

```python
x=dtset.iloc[:,:-1]
y=dtset["Hours Studied"]
```

```python
sc=StandardScaler()
sc.fit(x)
```

```python
arr=sc.transform(x)
arr
```

```python
df=pd.DataFrame(arr,columns=x.columns)
df
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression,Lasso,Ridge
```

```python
#checking using Linearregression
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.coef_
```

```python
plt.figure(figsize=(15,5))
plt.bar(x.columns,lr.coef_)
plt.title("linearRe")
plt.xlabel("columns")
plt.ylabel("coef") 
plt.show()
```

```python
la=Lasso(alpha=0.5)
la.fit(x_train,y_train)
```

```python
la.score(x_test,y_test)*100
```

```python
la.coef_
```

```python
plt.figure(figsize=(15,5))
plt.bar(x.columns,la.coef_)
plt.title("lasso")
plt.xlabel("columns")
plt.ylabel("coef") 
plt.show()
```

```python
#ridge
ri=Ridge(alpha=10)
ri.fit(x_train,y_train)
```

```python
ri.score(x_test,y_test)*100
```

```python
plt.figure(figsize=(15,5))
plt.bar(x.columns,ri.coef_)
plt.title("Ridge")
plt.xlabel("columns")
plt.ylabel("coef") 
plt.show()
```
