```python
df_country=pd.read_csv("exl.csv")
df_country
```

```python
df.columns
```

```python
final=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final.head(2)
```

```python
final.columns
```

```python
names=final.Currency.value_counts().index
names
```

```python
he=final.Currency.value_counts().values
he
```

```python
#pie chart
plt.pie(he[:3],labels=names[:3],autopct='%1.f%%',startangle=30)
plt.title("pie chart")
plt.show()
```

```python
final.columns
```

```python
ratings=final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize'] = (12, 6)
sn.barplot(x="Aggregate rating",y="Rating Count",data=ratings)
```

```python
sn.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
sn.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])

```
