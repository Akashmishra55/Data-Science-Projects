```python
import pandas as pd
```

```python
df1=pd.read_csv("student-scores.csv")
df=df1.copy()
df.head()
```

```python
df.drop(columns=["id","first_name","last_name","email"],axis=1,inplace=True)
```

```python
#create features from all score
df["total_score"] = df["math_score"] + df["history_score"] + df["physics_score"] + df["chemistry_score"] + df["biology_score"] + df["english_score"] + df["geography_score"]
df["average_score"] = df["total_score"] / 7
df.head()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
le.fit(df["gender"])
df["gender"]=le.transform(df["gender"])
       
```

```python
le=LabelEncoder()
le.fit(df["part_time_job"])
df["part_time_job"]=le.transform(df["part_time_job"])
```

```python
le=LabelEncoder()
le.fit(df["extracurricular_activities"])
df["extracurricular_activities"]=le.transform(df["extracurricular_activities"])
       
```

```python
le=LabelEncoder()
le.fit(df["career_aspiration"])
df["career_aspiration"]=le.transform(df["career_aspiration"])
       
```

```python
from imblearn.under_sampling import RandomUnderSampler
```

```python
df.head()
```

```python
df.columns
```

```python
column_to_move="career_aspiration"
last_position=len(df.columns)-1
columns=list(df.columns)
columns.remove(column_to_move)
columns.insert(last_position,column_to_move)
df=df[columns]
```

```python
df
```

```python
x=df.iloc[:,:-1]
```

```python
y=df["career_aspiration"]
```

```python
ru=RandomUnderSampler()
ru_x,ru_y=ru.fit_resample(x,y)
```

```python
ru_y.value_counts()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(ru_x,ru_y,test_size=0.2,random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
x_train_scaled=ss.fit_transform(x_train)
x_test_scaled=ss.transform(x_test)
```

```python
x_train_scaled
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score,classification_report,confusion_matrix
```

```python
lr=LogisticRegression()
lr.fit(x_train_scaled,y_train)
```

```python
y_pred=lr.predict(x_train_scaled)
```

```python
accuracy=accuracy_score(y_test,y_pred)
Classification=classification_report(y_test,y_pred)
confusion_matrix=confusion_matrix(y_test,y_pred)
```
